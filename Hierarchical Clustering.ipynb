```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
plt.rcParams['figure.figsize'] = (5,5)
plt.style.use('ggplot')
```

```python
data = pd.read_csv('movie_metadata1.csv')
f1 = data['budget'].values
f2 = data['gross'].values
```

```python
f1
```

```python
f2
```

```python
plt.scatter(f1,f2)
```

```python
fb = f1[0:10]
fg = f2[0:10]
```

```python
fb
```

```python
fg
```

```python
X = np.array(list(zip(f1, f2)))
```

```python
X
```

```python
plt.scatter(fb,fg)
```

```python
Z = linkage(X, 'ward')
fig = plt.figure(figsize = (5,5))
dn = dendrogram(Z)
plt.show()
```

```python
Z = linkage(X, 'single')
fig = plt.figure(figsize = (5,5))
dn = dendrogram(Z)
plt.show()
```

```python
Z = linkage(X, 'median')
fig = plt.figure(figsize = (5,5))
dn = dendrogram(Z)
plt.show()
```

```python
Z = linkage(X, 'average')
fig = plt.figure(figsize = (5,5))
dn = dendrogram(Z)
plt.show()
```

```python
Z = linkage(X, 'ward')
fig = plt.figure(figsize = (5,5))
dn = dendrogram(Z)
plt.show()


```

```python
max_d=(.5*10**10)
from scipy.cluster.hierarchy import fcluster
clusters = fcluster(Z, max_d, criterion='distance')
```

```python
clusters
```

```python
set(clusters)
```

```python
org_data=pd.DataFrame(X)
```

```python
org_data['Group']=clusters
```
